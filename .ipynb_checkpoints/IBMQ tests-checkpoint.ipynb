```python
theta = 0
shots=1e3
g= 10
```

```python
qr = QuantumRegister(10)
cr = ClassicalRegister(10)
circuit = QuantumCircuit(qr,cr)

circuit.ry(theta,0)
circuit.ch(1,0)

circuit.ry(theta,2)
circuit.ch(3,2)
circuit.h(2)
circuit.h(3)

circuit.ry(theta,4)
circuit.ch(5,4)
circuit.rx(np.pi/2,4)
circuit.rx(np.pi/2,5)

circuit.ry(theta,6)
circuit.ch(7,6)
circuit.h(6)

circuit.ry(theta,8)
circuit.ch(9,8)
circuit.h(8)


circuit.measure(qr,cr)
```

```python
circuit.draw(output='mpl')
```

```python
QC_type = 1
if QC_type == 0:
    backend = provider.get_backend('ibmq_16_melbourne') # Quantum Computer
else: 
    backend = Aer.get_backend('qasm_simulator') # Quantum Simulator
```

```python
result = execute(circuit, backend, shots=shots).result()
count = result.get_counts(circuit)
```

```python
def prob(block,qq,count,shots):
    P=0
    for k in list(count.keys()): 
        split = []
        for i in range(0, len(k), 2):
            split.append( k[i] + k[i+1] )
        if split[-(block+1)] == qq:
            P += count[k]
    return P / shots
```

```python
def Hamiltonian(theta,count,shots=1024,g=1):
    P_00 = prob(0,'00',count=count,shots=shots) 

    Pxx_00 = prob(1,'00',count=count,shots=shots)
    Pxx_11 = prob(1,'11',count=count,shots=shots)

    Pyy_00 = prob(2,'00',count=count,shots=shots)
    Pyy_11 = prob(2,'11',count=count,shots=shots)

    Pzx_00 = prob(3,'00',count=count,shots=shots)
    Pzx_11 = prob(3,'11',count=count,shots=shots)

    Pix_00 = prob(4,'00',count=count,shots=shots)
    Pix_10 = prob(4,'10',count=count,shots=shots)
    #hamiltonian
    H_E = 3*(1 - P_00)
    H_B = (1/2)*( 5*(Pxx_00 + Pxx_11) + 5*(Pix_00 + Pix_10) + 3*(Pzx_00 + Pzx_11) - 3*(Pyy_00 + Pyy_11) -5)
    H = (g**2 /2)*H_E - (1/(2*g**2))*H_B
    return H 
```

```python
Hamiltonian(theta,count=count,shots=shots,g=g)
```
