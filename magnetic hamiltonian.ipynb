```python
import numpy as np
import matplotlib.pyplot as plt
import scipy 
from sympy.physics.wigner import wigner_6j
from pandas import *

def dim(j):
    return 2*j + 1

def left_plaquette(jl_f,ql_f,ja_f,qr_f,jl_i,ql_i,ja_i,qr_i):
    if jl_i == jl_f:
        jl = jl_i
    else: return 0
        
    M = (-1)**(2*(2*ja_f+2*jl-ql_i-qr_i)) 
    
    N = dim(ja_i)*dim(ja_f) * np.sqrt( dim(ql_i)*dim(ql_f)*dim(qr_i)*dim(qr_f))
    
    try:
        w1 = wigner_6j(jl,ja_i,ql_i,0.5,ql_f,ja_f)**2
        w2 = wigner_6j(jl,ja_i,qr_i,0.5,qr_f,ja_f)**2
    except:
        return 0
    return M*N*w1*w2


def B(jl_f,ql_f,ja_f,qr_f,jl_i,ql_i,ja_i,qr_i):
    left = left_plaquette(jl_f,ql_f,ja_f,qr_f,jl_i,ql_i,ja_i,qr_i)
    right= left_plaquette(ja_f,qr_f,jl_f,ql_f,ja_i,qr_i,jl_i,ql_i) 
    return 2*( right + left )
```

```python
B(0,0,0,0,0,0,0,0)
```

```python
# creating the states.

nums = {}

def d2b(x):
    """Decimal to Binary with base 4"""
    return format(int(x), '04b')

for i in range(16):
    nums[i] = d2b(i)
    
```

```python
def b2v(x):
    """binary to variable for lattice state"""
    return [ int(char)/2 for char in x]
```

```python
for i in nums:
    print (nums[i])
```

```python
Z = np.zeros([16,16])
```

```python
for i in nums:
    for j in nums:
        Z[i,j] = B(*b2v(nums[i])+b2v(nums[j]))
```
