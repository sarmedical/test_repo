```python
import numpy as np
import matplotlib.pyplot as plt
import scipy 
from sympy.physics.wigner import wigner_6j
from pandas import *

def dim(j):
    return 2*j + 1

def plaquette(jl_f,ql_f,ja_f,qr_f,jr_f,jl_i,ql_i,ja_i,qr_i,jr_i):
    if jl_i == jl_f:
        jl = jl_i
    else: return 0
    
    if jr_i == jr_f:
        jr = jr_i
    else: return 0
    
    M = (-1)**(2*( 2*ja_f + jl + jr - ql_i - qr_i )) 
    
    N = dim(ja_i)*dim(ja_f) * np.sqrt( dim(ql_i)*dim(ql_f)*dim(qr_i)*dim(qr_f))
    
    try:
        w1 = wigner_6j(jl,ja_i,ql_i,0.5,ql_f,ja_f)**2
        w2 = wigner_6j(jr,ja_i,qr_i,0.5,qr_f,ja_f)**2
    except:
        return 0
    return M*N*w1*w2


def B(q1i,p1i,q2i,p2i,q3i,p3i,q4i,p4i,q1f,p1f,q2f,p2f,q3f,p3f,q4f,p4f):
    plaquette1 = plaquette(p4i,q1i,p1i,q2i,p2i,p4f,q1f,p1f,q2f,p2f)
    plaquette2 = plaquette(p1i,q2i,p2i,q3i,p3i,p1f,q2f,p2f,q3f,p3f) 
    plaquette3 = plaquette(p2i,q3i,p3i,q4i,p4i,p2f,q3f,p3f,q4f,p4f)
    plaquette4 = plaquette(p3i,q4i,p4i,q1i,p1i,p3f,q4f,p4f,q1f,p1f)
    return 2*( plaquette1 + plaquette2 + plaquette3 + plaquette4)
```
