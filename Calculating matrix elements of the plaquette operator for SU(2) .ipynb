```python
import numpy as np
import matplotlib.pyplot as plt
import scipy 
from sympy.physics.wigner import wigner_6j


def dim(j):
    return 2*j + 1

def matrix_element(jl_f,ql_f,ja_f,qr_f,jr_f,jl_i,ql_i,ja_i,qr_i,jr_i):
    if jl_i == jl_f:
        jl = jl_i
    else: return 7
        
    if jr_i == jr_f:
        jr = jr_i
    else: return 7
        
    M = (-1)**(2*(jl+jr+2*ja_f-ql_i-qr_i)) 
    
    N = dim(ja_i)*dim(ja_f) * np.sqrt( dim(ql_i)*dim(ql_f)*dim(qr_i)*dim(qr_f))
    
    w1 = wigner_6j(jl,ja_i,ql_i,0.5,ql_f,ja_f)
    w2 = wigner_6j(jl,ja_i,ql_i,0.5,ql_f,ja_f)
    w3 = wigner_6j(jr,ja_i,qr_i,0.5,qr_f,ja_f)
    w4 = wigner_6j(jr,ja_i,qr_i,0.5,qr_f,ja_f)
    
    return M*N*w1*w2*w3*w4
```

```python
matrix_element(0,0,0,0,0,0,0.5,0.5,0.5,0)
```
