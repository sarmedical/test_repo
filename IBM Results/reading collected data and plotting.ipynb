```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = open("g_5_results.txt",'r')
```

```python
k=f.readlines()
```

```python
k
```

```python
theta = []
H = []
```

```python
for i in k:
    j = i[:-1].split(',' )
    theta.append( float( j[0][1:]) )
    H.append( float( j[1][:-1]))
```

```python
plt.figure(figsize=[14,7])
plt.plot(theta,H)
plt.axhline(0,color='k')
plt.axvline(0,color='k')
plt.title('')
plt.show()
```
